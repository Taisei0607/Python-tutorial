````python
def my_max(arg1, arg2):
    if arg1 - arg2 > 0:
        return arg1
    elif arg1 -arg2 < 0:
        return arg2
    else:
        return("２つの引数は等しい")
````

## 上のコードへのコメント
- 引数名を読んだ人が引数が何を意味いるのか分からないから、引数名を引数の種類や型、意味するものに変更しよう (変数名についても同様)
- 関数が何をしているかわかりやすいように、`if 引数1 > 引数2`という条件式にした方がわかりやすい
- 2つの引数が等しいときの処理も書けているのはGood

````python
def my_max_2(num_1, num_2):
    if num_1 > num_2:
        return num_1
    elif num_1 < num_2:
        return num_2
    else:
        return("2つの引数は等しい")
````

````python
print(my_max(10,8), my_max(7,12), my_max(-3, -7))
````

````python
print(my_max(arg1=13, arg2=9), my_max(arg2=9, arg1=7))
````

````python
def hello(x):
    for x in list(range(x)):
        print("hello")
````

## 上のコメント
- `for x in list(range(x))`という書き方は冗長で、range関数はこの場合0~x-1のx個の連続した数値のオブジェクトなので`range(x)`だけで十分
- ここでも引数名

## 追加課題
- helloと出力される前に `1回目: hello`というように何回目のhelloなのか表示してください
  - 文字列の連結は使わないで書いてください。`str1 + str2`はNG

````python
def hello_2(num):
    for num in range(num):
        print(num + 1, "hello")

hello_2(10)
````

````python
class Counter:
    def __init__(self):
        self.char_counter = {}

    def add_text(self, text):
        for char in text:
            self.char_counter[char] = self.char_counter.get(char, 0) + 1
        return (self.char_counter)
    
    def get_counts(self, char):
        if char in self.char_counter.keys():
            count = self.char_counter[char]
            return count
        else:
            return None
````

````python
counter = Counter()
````

````python
counter.add_text('arigato')
````

````python
counter.get_counts('a')
````

````python
counter.get_counts('c')
````
