```python
def my_abs(x):
    if x < 0:
        x *= -1
        return x
    else:
        return x
```

```python
print(my_abs(-1), my_abs(3))
```

```python
def my_max(arg1, arg2):
    if arg1 - arg2 > 0:
        return arg1
    elif arg1 -arg2 < 0:
        return arg2
    else:
        return("２つの引数は等しい")
```

```python
print(my_max(10,8), my_max(7,12), my_max(-3, -7))
```

```python
print(my_max(arg1=13, arg2=9), my_max(arg2=9, arg1=7))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_gaussian_profile(ax, z_label, w_val, label_w):
    # データの生成(y軸方向の分布)
    y = np.linspace(-2.5, 2.5, 500)
    # ガウス分布:E = exp(-(y/w)^2)
    E = np.exp(-(y/w_val)**2)
    
    # メインの曲線を描画
    ax.plot(E, y, color = 'black', linewidth = 2)
    
    # 軸の設定
    ax.axhline(0, color = 'black', linewidth = 1) # 横軸(E方向)
    ax.axvline(0, color = 'black', linewidth = 1) # 縦軸
    
    # 1/eの垂直線と水平線の描画
    e_inv = np.exp(-1) # 約0.368
    ax.plot([e_inv, e_inv], [-w_val, w_val], color = 'black', linewidth = 1) # 垂直線
    ax.plot([0, e_inv], [w_val, w_val], color = 'black', linewidth = 1) # 上の水平線
    ax.plot([0, e_inv], [-w_val, -w_val], color = 'black', linewidth = 1) # 下の水平線
    
    # テキストラベル
    ax.set_title(z_label, loc = 'center', pad = 10)
    ax.text(e_inv+0.05, w_val+0.2, '1/e', fontsize = 10)
    ax.text(-0.1, 0.1, '0', fontsize = 10)
    ax.text(-0.6, -w_val, label_w, fontsize = 12)
    
    # グラフの見た目調整
    ax.set_xlim(-0.5, 1.2)
    ax.set_ylim(-2.5, 2.5)
    ax.axis('off') # 枠線を消す
    # 矢印を模した描画(簡易版)
    ax.annotate('', xy = (1.2, 0), xytext = (0, 0), arrowprops = dict(arrowstyle = "->"))
    ax.annotate('', xy = (0, 2.5), xytext = (0, -2.5), arrowprops = dict(arrowstyle = "->"))
    ax.text(1.1, 0.1, 'E')

# 描画開始
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 5))

# 左側: z=0 (ビームウエスト)
draw_gaussian_profile(ax1, 'z=0', w_val = 1.0, label_w = r'$\omega(z=0)$')

# 右側: z=z (広がった状態)
draw_gaussian_profile(ax2, 'z=z', w_val = 1.6, label_w = r'$\omega(z=z)$')

plt.tight_layout()
plt.show()
```
